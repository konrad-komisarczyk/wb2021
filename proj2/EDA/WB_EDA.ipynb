```python
from sklearn.model_selection import train_test_split
```

```python
X_e = df_expired.drop(['death'], axis = 1)
y_e = df_expired['death']
X_i = df_icu.drop(['ICU.or.not'], axis = 1)
y_i = df_icu['ICU.or.not']
```

```python
X_e_train, X_e_test, y_e_train, y_e_test = train_test_split(X_e, y_e, test_size = 0.2)
X_i_train, X_i_test, y_i_train, y_i_test = train_test_split(X_i, y_i, test_size = 0.2)
```

### Test Boruty

```python
from boruta import BorutaPy
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=-1, class_weight='balanced', max_depth=3)
feat_selector = BorutaPy(rf, n_estimators='auto', max_iter=100)
feat_selector.fit(X_e_train.values, y_e_train)
```

```python
feat_selector.ranking_
```

```python
feat_selector_icu = BorutaPy(rf, n_estimators='auto', max_iter=100)
feat_selector_icu.fit(X_i_train.values, y_i_train)
```

```python
X_i_train.columns[feat_selector_icu.support_]
```

```python
X_e_train.columns[feat_selector.support_]
```

```python
df_expired_selected = X_e.loc[:, feat_selector.support_]
df_icu_selected = X_i.loc[:, feat_selector_icu.support_]
```

### Sklejanie ramek i zapisywanie do csvek

```python
df_expired_selected['death'] = y_e
```

```python
df_icu_selected['ICU.or.not'] = y_i
```

```python
df_expired_selected.to_csv("expired_data(extra).csv")
```

```python
df_icu_selected.to_csv("icu_data(extra).csv")
```
